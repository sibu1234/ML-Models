```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
make_blobs
```

```python
plt.rcParams['figure.figsize']=(16,9)
X,y=make_blobs(n_samples=800, n_features=3,centers=2)
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2])
```

```python
X
```

```python
kmeans1=KMeans(n_clusters=5)
```

```python
kmeans1=kmeans1.fit(X)
```

```python
labels=kmeans1.predict(X)
```

```python
labels
```

```python
Centroidarray=kmeans1.cluster_centers_
```

```python
Centroidarray
```

```python
fig=plt.figure()
ax=Axes3D(fig)
ax.scatter(X[:,0],X[:,1],X[:,2],c='y')
ax.scatter(Centroidarray[:,0],Centroidarray[:,1],Centroidarray[:,2],marker='*',c='r',s=1000)
```
